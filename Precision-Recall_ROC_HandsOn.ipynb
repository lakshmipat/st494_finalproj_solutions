```python
#Load the Iris dataset and convert it into a binary classification problem:

import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Combine two of the classes to create a binary classification problem
y_binary = np.where(y == 2, 1, 0)
```

```python
#Split the dataset into training and testing sets:

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=42)
```

```python
#Train a logistic regression model on the training data:

from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=42)
classifier.fit(X_train, y_train)
```

```python
#Generate ROC and PR curves:
from sklearn.metrics import plot_roc_curve, plot_precision_recall_curve

roc_plot = plot_roc_curve(classifier, X_test, y_test)
pr_plot = plot_precision_recall_curve(classifier, X_test, y_test)
```

```python
#Compare the AUC values:

roc_auc = roc_plot.roc_auc
pr_auc = pr_plot.average_precision

print("ROC AUC:", roc_auc)
print("PR AUC:", pr_auc)
```
